### Zero Shot Classification

```python
model = pipeline("zero-shot-classification")
```

```python
model("This is a course on Huggingface", candidate_labels=["politics", "education", "sports", "business"])
```

```python
model("RELIANCE stocks rose y 1.2% yesterday", candidate_labels=["politics", "education", "sports", "business"])
```

### Using a specific model

```python
model = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M-Instruct")
```

```python
text = model("Use 100 to complete the paragraph: Gravity is a force that .... ")
```

```python
print(text[0]['generated_text'])
```

```python
model('classify the statement into ["politics", "education", "sports", "business"]: This is a course on AI')
```

### Image Classification

```python
url = r"https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/pipeline-cat-chonk.jpeg"
```

```python
image_classifier = pipeline(
    task="image-classification",
    model="google/vit-base-patch16-224"
)
```

```python
image_classifier(url)
```

```python
image_classifier(r"https://images.pexels.com/photos/112460/pexels-photo-112460.jpeg?cs=srgb&dl=car-vehicle-luxury-112460.jpg&fm=jpg")
```
